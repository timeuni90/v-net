```python
import matplotlib.pyplot as plt
import SimpleITK as sitk
path = "E:\\code\\pycharm-workspace\\dataset\\" \
       "PROMISE12\\TrainingData_Part1\\Case02_segmentation.mhd"
image = sitk.ReadImage(path)
images_seg = sitk.GetArrayFromImage(image)
print(images_seg.shape)
path = "E:\\code\\pycharm-workspace\\dataset\\" \
       "PROMISE12\\TrainingData_Part1\\Case02.mhd"
image = sitk.ReadImage(path)
images = sitk.GetArrayFromImage(image)

def plt_itk(images):
    length = images.shape[0]
    col = 2
    row = length // col
    if length % col != 0:
        row += 1
    figure, axs = plt.subplots(row, col, figsize=(col*5, row*5))
    for i in range(row):
        for j in range(col):
            index = i * col + j
            if index < length:
                axs[i, j].imshow(images[i], cmap=plt.cm.bone)
                axs[i, j].axis("off")

def compare_itk(imgs_1, imgs_2):
    length = len(imgs_1)
    figure, axs = plt.subplots(length, 2, figsize=(2 * 5, length * 5))
    for i in range(length):
        for j in range(2):
            plot = axs[i, j]
            plot.axis("off")
            if j == 1:
                plot.imshow(imgs_2[i], cmap=plt.cm.bone)
                continue
            plot.imshow(imgs_1[i], cmap=plt.cm.bone)

compare_itk(images, images_seg)
```
